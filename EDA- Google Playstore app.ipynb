### 4.2.2. Convert the whole Size Column into bytes
- Lets define function to convert the ***Size*** to ```Numeric``` 

````python
# Convert 'Size' to numeric, 
def convert_size(Size):
    if isinstance(Size, str):
        if 'k' in Size:
            return float(Size.replace('k', "")) *1024  # cinverting 'k' to bytes
        elif 'M' in Size:
            return float(Size.replace('M', "")) * 1024 * 1024 # converting 'M' to bytes
        elif 'G' in Size:
            return float(Size.replace('G', "")) * 1024 * 1024 * 1024 # converting 'G' to bytes
        elif 'Varies with device' in Size:
            return np.nan
    return Size 
````

- Applying the function Column ***Size***

````python
df['Size'] = df['Size'].apply(convert_size)
````

### 5.1. Numerical Features

````python
# Plotting the distribution of app ratings
plt.figure(figsize=(12, 6))
sns.histplot(df['Rating'], bins=10, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.axvline(df['Rating'].mean(), color='red', linestyle='--', label='Mean Rating')
plt.axvline(df['Rating'].median(), color='green', linestyle='--', label='Median Rating')
plt.legend()
plt.show()
````

````python
# Draw the plot of Price with log for better visualization
plt.figure(figsize=(10, 6))
sns.histplot(np.log1p(df['Price'].replace(0, np.nan)), bins=30, kde=True)
plt.title('Distribution of App Prices (Log Scale)')
plt.xlabel('Log(Price + 1)')
plt.ylabel('Frequency')
plt.axvline(df['Price'].mean(), color='red', linestyle='--', label='Mean Rating')
plt.axvline(df['Price'].median(), color='green', linestyle='--', label='Median Rating')
plt.show()
````

- Identify top-performing developers.

````python
# Group by Developer Id and calculate the average installs and ratings
average_per_developer = df.groupby('Developer Id').agg(
    average_installs=('Installs', 'mean'),
    average_rating=('Rating', 'mean')
).reset_index()

# Sort by average installs and ratings
top_developers_installs = average_per_developer.sort_values(by='average_installs', ascending=False)
top_developers_ratings = average_per_developer.sort_values(by='average_rating', ascending=False)

# Display top developers by average installs
print("Top Developers by Average Installs:")
print(top_developers_installs.head())

# Display top developers by average ratings
print("\nTop Developers by Average Ratings:")
print(top_developers_ratings.head())
````

- Content rating (e.g., Everyone, Teen, Mature) affect app installs and ratings
- Study how age targeting affects success.

````python
# Group by Content Rating and calculate the average installs and ratings
content_rating_performance = df.groupby('Content Rating').agg(
    average_installs=('Installs', 'mean'),
    average_rating=('Rating', 'mean')
).reset_index()

# Sort by average installs for better insights
content_rating_performance = content_rating_performance.sort_values(by='average_installs', ascending=False)

# Display the results
print("Average Installs and Ratings by Content Rating:")
print(content_rating_performance)
````
